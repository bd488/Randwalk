## Identifying the location of pulse and fitting to Gauss

```python
import glob
path = "*.csv"
for fname in glob.glob(path):
    #inserting data
    sequence,  CH1, CH2  = np.loadtxt(fname,
                                  usecols=(0,1,2),
                                  skiprows=502,
                                  delimiter=',',
                                  unpack=True)
#determining where the pulse satrts and how much time passes each increment is
    start, increment  = np.loadtxt(fname, usecols=(3,4), skiprows=1, max_rows=1, delimiter=',', unpack=True)

   

#time
    time = sequence*increment

#time from start of pulse
    timep = time+start

#plot graph
    plt.plot(time, CH2, "k.")
    plt.xlabel('time (s)')
    plt.ylabel('Voltage (V)')
    plt.title(fname)

    plt.show()
    
    x = time
    y = CH2

    point = max(y)
    print("the approximate highest peak is: " +str(point))

    estimate=np.median(np.where(y==point))
    xtimate=np.median(x[np.where(y==point)])

    print("Pulse started at:")
    print(start)
    print("Each time division is")
    print(increment)

    print("the approximate centre of peak is found at value: " +str(estimate))
    print("this has the x value of: " +str(xtimate))
    
    sequencenew,  CH1new, CH2new  = np.loadtxt(fname,
                                  usecols=(0,1,2),
                                  skiprows=(502 + int(estimate)-202),
                                  max_rows=400,
                                  delimiter=',',
                                  unpack=True)

#timenew for this intervall
    timenew= sequencenew*increment

#plot graph of only pulse
    plt.plot(timenew, CH2new, "k.")
    plt.xlabel('time (s)')
    plt.ylabel('Voltage (V)')
    plt.title(fname)

    plt.show()
    
    xn = timenew
    yn = CH2new

    n = len(xn)                          #length of data
    mean = np.mean(xn)                  # mean of x
    sigma = np.std(xn)                  # standard deviation of x

    def gaus(xn,a,xn0,sigma):                      #run gauss
        return a*exp(-(xn-xn0)**2/(2*sigma**2))

    popt,pcov = curve_fit(gaus,xn,yn,p0=[1,mean,sigma])

    plt.plot(xn,yn,'b+:',label='data')
    plt.plot(xn,gaus(xn,*popt),'ro:',label='fit')
    plt.legend()
    plt.title(fname)
    plt.xlabel('Time (s)')
    plt.ylabel('Voltage (V)')
    plt.show()

    print("Final y peak value: " + str(np.max([yn])))
    print("Final x value: " + str(mean))
    print("-------------------------------------------------------------------------------------")
```
